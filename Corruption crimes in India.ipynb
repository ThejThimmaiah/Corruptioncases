# Bar graph showing the Total cases with respect to states/UT

```python
crime.groupby("Abbreviation").Total_cases.mean().plot(kind="bar")
```

```python
plt.hist(crime["Total_cases"])
plt.title("Total no of cases")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid()
plt.show()
```

```python
crime.groupby("Abbreviation").Total_cases.mean().plot(kind="bar")
plt.title("Total cases")
plt.ylabel("cases")
plt.xlabel("count")
plt.yticks(rotation=90)
plt.show()
```

# Pie chart showing the percentage of Total cases with respect to states/UT

```python
plt.figure(figsize=(15,15))
crime.groupby('Abbreviation')['Total_cases'].nlargest(15).plot(kind='pie',autopct="%1.2f%%")
```
